```python
# Convert dictionary into dataframe

#for key in dictionary_of_articles.keys():
    #print key, dictionary_of_articles[key]

articles_df = pd.DataFrame({"article_source":dictionary_of_articles.keys(),
                           "article_text": dictionary_of_articles.values(),
                           })
articles_df
articles_df["article_word_count_unique"] = "empty"
articles_df["article_word_count"] = "empty"
articles_df["polarity"] = "empty"
articles_df["noun_phrases"] = "empty"
articles_df["subjectivity"] = "empty"
articles_df["tag_breakdown"] = "empty"

for i in range(len(articles_df["article_text"])):
    text = articles_df["article_text"][i]
    list_of_words = text.split()
    articles_df["article_word_count"][i] = len(list_of_words)
    articles_df["article_word_count_unique"][i] =len(set(list_of_words))
    
    ######    ######    ######    ######    ######
    #Getting Textblob specific features
    article_blob = TextBlob(text)
    articles_df["polarity"][i] = article_blob.polarity
    articles_df["noun_phrases"][i] = article_blob.noun_phrases
    articles_df["subjectivity"][i] = article_blob.sentiment.subjectivity
    
    #Splits up words into it's tags: e.g nnp, nn, dt...
    # Create a dictionary of counts -- how many words are nnp, nn, etc.
    tags_dictionary = defaultdict(int)
    for word in article_blob.tags:
        tags_dictionary[word[1]] +=1
    
    # Converts that dictionary into percentage for the whole article
    
    for tag in tags_dictionary.keys():
        tags_dictionary[tag] = round(tags_dictionary[tag]/float(articles_df["article_word_count"][i]),3)

    articles_df["tag_breakdown"][i] = tags_dictionary.items()
```
